```python
#Smart real estate predictor using liner Regression

#creating a dataset using dictionery

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression


data={
    "Area":[1000 ,2000, 3000, 5000,6000],
    "price":[100,200,400,600,800]
}

print(data)

#converting to dataframe so we can reshape  data in 2 d arrray to train model

df=pd.DataFrame(data)

print(df)

#preparing data in x and Y to train AI Model
x=df[['Area']].values #feature iput #double bracket so we can create 2 d array it will b in column
y=df['price'].values   #features as output
print(x)
print(y)

#Brain: create and tain AI
model = LinearRegression()
model.fit(x, y) # The AI "studies" the relationship between Area and Price

# What would a 2200 sq ft house cost?
new_house_area = np.array([[2200]])
prediction = model.predict(new_house_area)

print(f"Prediction for a 2200 sq ft house: ${prediction[0]:.2f}k")

#  VISUALIZE: See the "AI Logic"
plt.scatter(x, y, color='blue', label='Actual Sales') # The dots
plt.plot(x, model.predict(x), color='red', label='AI Prediction Line') # The line
plt.xlabel('Square Feet')
plt.ylabel('Price ($1000s)')
plt.legend()
plt.show()
```
