```python
#Create an Agentic AI tool to write description of house and create instagram post using crewAI
import os
from crewai import Agent, Task ,Crew
import requests
from langchain_google_genai import ChatGoogleGenerativeAI
# 1. Install the necessary libraries in Colab
#!pip install -q crewai langchain-google-genai
from google.colab import userdata



#LLM set up
os.environ["GOOGLE_API_KEY"] = userdata.get('GoogleKey')
#  Setup Brain
llm = ChatGoogleGenerativeAI(model="gemini-2.5-flash")

#  Define the Team (Agents)
writer = Agent(
  role='Professional Biographer',
  goal='Summarize house details into a warm, inviting 2-paragraph description.',
  backstory='You specialize in emotional storytelling for real estate.',
  llm=llm
)

social_manager = Agent(
  role='Social Media Strategist',
  goal='Convert long descriptions into viral Instagram posts.',
  backstory='You are an expert at using emojis and hashtags to get engagement.',
  llm=llm
)

#  Define the Workflow (Tasks)
house_data = "Area: 2200 sq ft, 3 Bedrooms, 8 years old, Price: $395,000"

task1 = Task(
  description=f"Write a formal listing for this house: {house_data}.",
  expected_output="A 2-paragraph professional description.",
  agent=writer
)

task2 = Task(
  description="Take the formal listing and turn it into a fun Instagram post.",
  expected_output="An Instagram post with emojis and 5 relevant hashtags.",
  agent=social_manager,
  context=[task1] # This tells the agent to wait for the writer to finish!
)

# The Crew (The Team)
my_team = Crew(
  agents=[writer, social_manager],
  tasks=[task1, task2],
  verbose=True # This lets you "see" the agents talking to each other in the console
)

result = my_team.kickoff()

print("\n--- FINAL SOCIAL MEDIA POST ---")
print(result)
```
