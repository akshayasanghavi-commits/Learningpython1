```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#Student garade analytic system Project

#Getting data from User storing it in list ,creating tuple

print("How many students do you have :")
Numberofstudents=int(input())
print(f"Total Number of students is {Numberofstudents}")

x=1
studentslist=[]
subjecttuple=("Maths","Scienec","English")
grades=np.zeros((Numberofstudents,len(subjecttuple)),dtype=int)


while x <= Numberofstudents :
  print(f"Enter{x} student Name ")
  StudentName=input()
  studentslist.append(StudentName)

  for y in range (len(subjecttuple)):
     print(f"add {subjecttuple[y]}subject grades")
     z=int(input())
     grades[x-1,y]=z

  print(grades)
  x=x+1

print(studentslist)

##now convert all dats into table using panda
df=pd.DataFrame(grades,index=studentslist,columns=subjecttuple)
print (df)

df['Average'] = df.mean(axis=1) # Calculate mean for each student
df['Status'] = np.where(df['Average'] >= 60, "Pass", "Fail")

print("--- Student Grade Report ---")
print(df)


#use of matplotlib to create chart

df_sorted=df.sort_values( by='Average', ascending=False)

# use plt.bar to create a bar chart
plt.bar(df_sorted.index, df_sorted['Average'], color='skyblue')

# Adding "Meta-data" to the chart (Labels and Titles)
plt.xlabel('Students')
plt.ylabel('Average Grade')
plt.title('Class Performance: Student Average Grades')
plt.ylim(0, 100) # Ensure the scale goes from 0 to 100

# This loop adds the exact number on top of each bar
for i, value in enumerate(df_sorted['Average']):
    plt.text(i, value + 1, f'{value:.1f}', ha='center')

# Display the chart
plt.show()
```
